#### Cara Import Module dan Mengecek Version

#### Mempersiapkan Data Frame

Kita akan mempersiapkan data frame yang terdiri dari 40 baris dan 5 kolom. Untuk membentuk object data frame, kita bisa menggunakan perintah `pd.DataFrame()`.

```python
n_rows = 40 #Ini merepresentasikan jumlah baris yaitu 40 baris.
n_cols = 5 #Ini merepresentasikan jumlah kolom yaitu 5 kolom.
cols = tuple('ABCDE') #Setiap kolom akan diberi label A, B, C, D, dan E.

#Dalam data frame ini akan terisi dengan angka random (Integer) dari 1-20.
#Jumlah baris dan kolom sesuai dengan yang sudah ditentukan diatas.
df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), 
                  columns=cols)

# Menampilkan 5 baris pertama.
df.head()
```

#### Line Plot

Untuk melakukan visualisasi data sebagai line plot, kita bisa menggunakan method `.plot` dan memberikan parameter kind dengan nilai `line`. Pada hasil plotting dibawah:
- Garis merepresentasikan setiap kolom.
- Sumbu x merepresentasikan baris.
- Sumbu y merepresentasikan nilai dalam data frame.

```python
# Mereview isi data frame (5 baris pertama).
df.head()
```

```python
# Melakukan plotting untuk semua kolom.
df.plot(kind='line')
```

```python
# Melakukan plotting untuk kolom A dan B saja.
df[['A', 'B']].plot(kind='line')
```

#### Bar Plot

Untuk melakukan visualisasi data sebagai bar plot, kita bisa menggunakan method `.plot` dan memberikan parameter kind dengan nilai `bar`. Pada hasil plotting dibawah:
- Sumbu x merepresentasikan baris.
- Sumbu y merepresentasikan nilai dalam data frame.

```python
# Mereview isi data frame (5 baris pertama).
df.head()
```

```python
# Melakukan plotting untuk semua kolom.
df.plot(kind='bar')
```

```python
# Melakukan plotting untuk kolom A dan B saja.
df[['A', 'B']].plot(kind='bar')

# Kita bisa melakukan plotting untuk kolom A dan B pada 5 baris pertama saja dengan menambahkan .head() sebelum .plot().
# df[['A', 'B']].head().plot(kind='bar')
```

```python
# Melakukan plotting untuk kolom A dan B pada 5 baris pertama saja dengan posisi plotting horizontal.
df[['A', 'B']].head().plot(kind='barh')
```

#### Area Plot

```python
# Mereview isi data frame (5 baris pertama).
df.head()
```

```python
# Melakukan plotting untuk semua kolom.
df.plot(kind='area')
```

#### Scatter Plot

Untuk melakukan visualisasi data sebagai KDE, kita bisa menggunakan method `.plot`, menspesifikasi kolom yang ingin diperhatikan (harus 2 kolom), dan memberikan parameter kind dengan nilai `scatter`.
