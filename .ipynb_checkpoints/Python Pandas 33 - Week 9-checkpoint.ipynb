#### Mempersiapkan Data Frame

Kita akan menyiapkan data frame yang terdiri dari 2 kolom yaitu kolom pemain dan kolom goal. 

```python
d = {'pemain':['Budi', 'Joni', 'Iwan', 'Budi', 'Budi', 'Iwan', 'Asep', "Joni"],
     'goal':[2, 1, 3, 1, 1, 2, 2, 3]}

df = pd.DataFrame(d)
df
```

#### Menampilkan Nilai Kumulatif (Cummulative)

Terdapat beberapa fungsi yang digunakan untuk menampilkan nilai kumulatif:
- Fungsi `.cumsum()` berfungsi untuk mendata jumlah gol secara kumulatif. Contohnya total jumlah goal dari index ke-0 sampai ke-5 adalah 10.
- Kita juga bisa mengkombinasikan fungsi `.groupby()` dan `.cumsum()`. Contohnya kita mau mencari total jumlah goal untuk pemain bernama Budi yaitu 4.
- Fungsi `cummax()` berfungsi untuk mendata nilai tertinggi yang diperoleh sampai dengan titik tertentu.
- Fungsi `cummin()` berfungsi untuk mendata nilai terendah yang diperoleh sampai dengan titik tertentu.
- Fungsi `cumprof()`berfungsi untuk mendata hasil total perkalian kumulatif.

```python
# Menerapkan nilai kumulatif cumsum di kolom gol.
df['goal'].cumsum().to_frame()
```

```python
# Menampung nilai kumulatif sum dari kolom goal. 
df['jumlah_goal_kumulatif'] = df['goal'].cumsum()
df
```

```python
# Menerapkan kumulatif sum untuk setiap kelompok pemain
df['jumlah_goal_kumulatif_tiap_pemain'] = df.groupby('pemain')['goal'].cumsum()
df
```

```python
# Kumulatif max 
df['cummax'] = df['goal'].cummax()
df
```

```python
# Kumulatif min
df['cummin'] = df['goal'].cummin()
df
```

```python
# Kumulatif product
df['cumprod'] = df['goal'].cumprod()
df
```
