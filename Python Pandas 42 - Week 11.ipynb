#### Mempersiapkan Data Frame

Kali ini kita akan membentuk data frame menggunakan data frame classic yaitu titanic data set. Titanic data set adalah public data set yang bisa diakses dan digunakan secara bebas. Caranya adalah load data set titanic yang dalam format .csv menjadi format pandas data frame menggunakan method `pd.read_csv` dan setelah itu diarahkan ke lokasi dari file `titanicfull.csv`.

```python
# Kita akan menyimpan object data framenya ke dalam variabel df.
df = pd.read_csv('./data/titanicfull.csv')

# df.head digunakan untuk melihat 5 baris pertama dari df.
df.head()
```

#### Mengurutkan Data Berdasarkan Kolom Tertentu

Kita bisa mengurutkan data frame berdasarkan kolom tertentu dengan memanfaatkan method `sort_values()`. Jika kita ingin mengurutkan data dari besar ke kecil, kita bisa menambahkan parameter `ascending` dengan nilai `False`. Jika kita ingin mengurutkan data dari kecil ke besar, kita tidak perlu menambahkan parameter `ascending` lagi. Kita bisa menggunakan List jika mempunyai lebih dari 1 kolom yang akan digunakan sebagai acuan. Urutan penulisan elemen dalam List sangatlah penting karena berpengaruh dalam prioritas pengurutan data. 

```python
# Mengurutkan data frame dengan acuan kolom age dan hanya menampilkan 5 data pertama saja.
df.sort_values('age').head()
```

```python
# Mengurutkan data frame dengan acuan kolom age dari besar ke kecil dan hanya menampilkan 5 data pertama saja.
df.sort_values('age', ascending=False).head()
```

```python
# Mengurutkan data frame dengan acuan kolom survived dan kolom age, serta hanya menampilkan 5 data pertama saja.
df.sort_values(['survived', 'age']).head()
```
